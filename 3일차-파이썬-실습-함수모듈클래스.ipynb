### [실습문제] 산술연산 사용자 정의 함수 만들기


```python
# 방법1
def calc(n1, n2, op):
    if op == '+':
        result = n1 + n2
    elif op == '-':
        result = n1 - n2
    elif op == '*':
        result = n1 * n2
    elif op == '/':
        result = n1 / n2
    elif op == '//':
        result = n1 // n2
    elif op == '%':
        result = n1 % n2
    elif op == '**':
        result = n1 ** n2
    else:
        print('잘못된 기호')
        result = None
    
    return result

num1 = int(input('첫번째 숫자: '))
num2 = int(input('두번째 숫자: '))
op = input('연산 기호: ')

print(f'결과: {calc(num1, num2, op)}')
```

```python
# 방법2
def calc(n1, n2, op):
    if op in ['+','-','*','/','//','%','**']:
        return eval(n1 + op + n2)    
    else:
        return print('잘못된 기호')

num1 = input('첫번째 숫자: ')
num2 = input('두번째 숫자: ')
op = input('연산 기호: ')

print(f'결과: {calc(num1, num2, op)}')
```

### [실습문제] 암스트롱수 판별 함수 만들기 

```python
def is_arm(N):
    _1 = str(N)[0]
    _2 = str(N)[1]
    _3 = str(N)[2]
    
    if N == (int(_1)**3 + int(_2)**3 + int(_3)**3 ):
        return True    
    return False
    
for i in range(100, 1000):
    if is_arm(i):
        print(f'[{i}] is armstrong number!')

print('end---------')
```
